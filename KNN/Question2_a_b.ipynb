```python
#Student ID 1002080520

import numpy as np
from numpy import loadtxt
import csv

#CODE TO TAKE THE DATA AND STORE IN THE VARIABLE

file = open('q2_a_training.csv')
csvreader = csv.reader(file, delimiter = ",")
data = []
for row in csvreader:
    data.append(row)

train_data = []
for i in range(0 , len(data)):
    col= []
    for j in range(0,len(data[i])):
        if j==len(data[i])-1:
            col.append(data[i][j])
        else:
            col.append(float(data[i][j]))
    train_data.append(col)
file.close()
file = open('q2_a_test_data.csv')
csvreader = csv.reader(file , delimiter = ",")
t_data = []
for row in csvreader:
    t_data.append(row)
test_data = []
for i in range(0 , len(t_data)):
    col= []
    for j in range(0,len(t_data[i])):
            col.append(float(t_data[i][j]))
    test_data.append(col)
    
print("\n\nTraining Data \n\n")
print(train_data)
print("\n\nTest Data \n\n")
print(test_data)

#We have successfully imported the data
```

```python
def bubblesort(arr):   
    n = len(arr)
    # For loop to traverse through all
    # element in an array
    for i in range(n):
        for j in range(0, n - i - 1):
             
            # Range of the array is from 0 to n-i-1
            # Swap the elements if the element found
            #is greater than the adjacent element
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]


#NOW WE HAVE STORED THE TRAINING DATA IN train_data AND TEST DATA IN test_data

#Now while knowing the train data we will now find the cartesian distance between the training and test data

Distance = [] #Array of Distance & Label
for i in range(0 , len(test_data)):
    col = []
    for j in range(0,len(train_data)):
        dist_label =[]
        dist_label.append(train_data[j][4])
        X = test_data[i][0]
        Y = test_data[i][1]
        Z = test_data[i][2]
        W = test_data[i][3]
        x = train_data[j][0]
        y = train_data[j][1]
        z = train_data[j][2]
        w = train_data[j][3]
        dist = np.sqrt((X-x)**2 + (Y-y)**2 + (Z-z)**2 + (W-w)**2)
        dist_label.append(dist)
        col.append(dist_label)
    Distance.append(col)
print("\n DISTANCE MATRIX \n")
print(Distance)
print("\n")

#We here successfully calculated the cartesian distance and stored it in the Distance array 
#Now we need to find the minimum distance from every row and the corresponding label through which we can predict the given test data
```

```python
#We will now extract the distance measurement from Distance array and sort it in ascending order so that
#We can get the k nearest labels
only_dist =[]
for i in range(0, len(Distance)):
    cols=[]
    for j in range(0 , len(Distance[i])):
        cols.append(Distance[i][j][1])
    bubblesort(cols)
    only_dist.append(cols)
print(only_dist)

#We allocate the distance measured to the labels in ascending order

data_dist_label=[]
for i in range(0 , len(only_dist)):
    cols=[]
    for j in range(0 , len(only_dist[i])):
        for h in range(0 , len(only_dist[i])):
            if only_dist[i][j]==Distance[i][h][1]:
                d = Distance[i][h][0]
                break
        cols.append(d)
    data_dist_label.append(cols)

print(data_dist_label)

#We have successfully generated the distance label in ascending order
```

```python
#We will now find the K Nearest Neighbours and predict
a = " Metal"
b = " Plastic"
c = " Ceramic"

p_label_1=[]
p_label_3=[]
p_label_5=[]
p_label=[]

#Predicted Label for k=1

for i in range(0 , len(data_dist_label)):
    p_label_1.append(data_dist_label[i][0])

#Predicted Label for k=3
for i in range(0 , len(data_dist_label)):
    x=0
    y=0
    z=0
    cols=[]
    for j in range(0 , 3):
        if (str(data_dist_label[i][j]) == a):
            x=x+1
        if(str(data_dist_label[i][j])==b):
            y=y+1
        if(str(data_dist_label[i][j])==c):
            z=z+1
    if (x >= y) and (y >= z):
           largest = a
    elif (y >= x) and (y >= z):
        largest = b
    else:
        largest = c
    cols.append(largest)
    p_label_3.append(cols)

#Predicted Label for k=5
for i in range(0 , len(data_dist_label)):
    x=0
    y=0
    z=0
    cols=[]
    for j in range(0 , 5):
        if (str(data_dist_label[i][j]) == a):
            x=x+1
        if(str(data_dist_label[i][j])==b):
            y=y+1
        if(str(data_dist_label[i][j])==c):
            z=z+1
    if (x >= y) and (y >= z):
           largest = a
    elif (y >= x) and (y >= z):
        largest = b
    else:
        largest = c
    cols.append(largest)
    p_label_5.append(cols)
# We have now the predicted labels we will now append it to the test_matrix
predicted_output_1 = []
predicted_output_3=[]
predicted_output_5=[]
temp = 0
for i in range (0 , len(test_data)):
    cols_3=[]
    cols_1=[]
    cols_5=[]
    for j in range(0 , len(test_data[i])):
        cols_1.append(test_data[i][j])
        cols_3.append(test_data[i][j])
        cols_5.append(test_data[i][j])
    cols_3.append(p_label_3[i])
    cols_1.append(p_label_1[i])
    cols_5.append(p_label_5[i])
    predicted_output_1.append(cols_1)
    predicted_output_3.append(cols_3)
    predicted_output_5.append(cols_5)

#Hence we finally predixt and append the label to the data from the given input for KNN model with 1,3,5 nearest neighbours
print("\n\nThe prediction with 1 nearest neighbour\n\n")
print(predicted_output_1)
print("\n\nThe prediction with 3 nearest neighbour \n\n")
print(predicted_output_3)
print("\n\nThe prediction with 5 nearest neighbour \n\n")
print(predicted_output_5)
```
